```python
import boto3
import pickle
import numpy as np
from PIL import Image
import joblib
```

```python
s3 = boto3.client('s3', region_name='us-east-1')
model_file_name = 'credit_card_client_dataset.pkl'
bucket_name = 'geodykexp-ml-bucket'
```

```python
local_model_path = 'credit_card_client_dataset.pkl'
```

```python
# Upload the model file to S3
s3.upload_file(local_model_path, bucket_name, model_file_name)
```

```python
# Create an AWS Lambda function
lambda_client = boto3.client('lambda')
```

```python
# Specify the Lambda function details
function_name = 'credit_score_model'
runtime = 'Python 3.14'
handler = 'lambda_function.lambda_handler'
role = 'arn:aws:iam::255653206407:role/service-role/my_credit_score_model'
```

```python
# Create the Lambda function
lambda_client.create_function(
    FunctionName=function_name,
    Runtime=runtime,
    Handler=handler,
    Role=role,
    Code={
        'S3Bucket': bucket_name,
        'S3Key': model_file_name
    }
)
```
