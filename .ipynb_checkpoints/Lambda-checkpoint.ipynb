```python
import boto3
import pickle
import os
import json
import numpy as np
from PIL import Image
import joblib


s3 = boto3.client('s3', region_name='us-east-1')
model_file_name = 'credit_card_client_dataset'
bucket_name = 'geodykexp-ml-bucket'

local_model_path = f'/tmp/{model_file_name}'

# Upload the model file to S3
s3.upload_file(local_model_path, bucket_name, model_file_name)

# Create an AWS Lambda function
lambda_client = boto3.client('lambda')

# Specify the Lambda function details
function_name = 'your_lambda_function_name'
runtime = 'python3.8'
handler = 'your_lambda_handler_function'
role = 'arn:aws:iam::your_account_id:role/your_lambda_role'

# Create the Lambda function
lambda_client.create_function(
    FunctionName=function_name,
    Runtime=runtime,
    Handler=handler,
    Role=role,
    Code={
        'S3Bucket': bucket_name,
        'S3Key': model_file_name
    }
)


```
