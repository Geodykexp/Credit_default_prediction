```python
def plot_all_distributions(df, figsize=(20, 60), cols_per_row=5):
    """
    Plot histograms for all numeric columns in a grid layout
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = len(numeric_cols)
    n_rows = int(np.ceil(n_cols / cols_per_row))
    
    fig, axes = plt.subplots(n_rows, cols_per_row, figsize=figsize)
    axes = axes.flatten()
    
    for idx, col in enumerate(numeric_cols):
        ax = axes[idx]
        ax.hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
        ax.set_title(f'{col}', fontsize=10)
        ax.set_xlabel('Value', fontsize=8)
        ax.set_ylabel('Frequency', fontsize=8)
        ax.tick_params(labelsize=7)
    
    # Hide empty subplots
    for idx in range(n_cols, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('all_distributions.png', dpi=150, bbox_inches='tight')
    plt.show()

# Method 2: Statistical summary with visualizations
def plot_distributions_with_stats(df, save_path='distribution_report.png'):
    """
    Create a more compact visualization with key statistics
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    # Calculate statistics
    stats = df[numeric_cols].describe().T
    stats['skewness'] = df[numeric_cols].skew()
    stats['kurtosis'] = df[numeric_cols].kurtosis()
    
    print("Statistical Summary:")
    print(stats)
    
    # Plot distributions grouped by similar scales
    fig = plt.figure(figsize=(20, 60))
    cols_per_row = 5
    n_rows = int(np.ceil(len(numeric_cols) / cols_per_row))
    
    for idx, col in enumerate(numeric_cols, 1):
        plt.subplot(n_rows, cols_per_row, idx)
        
        # Histogram with KDE
        plt.hist(df[col].dropna(), bins=30, alpha=0.7, edgecolor='black', density=True)
        
        # Add KDE line if there's enough variation
        if df[col].std() > 0:
            df[col].dropna().plot(kind='kde', ax=plt.gca(), color='red', linewidth=2)
        
        plt.title(f'{col}\nMean: {df[col].mean():.2f}, Std: {df[col].std():.2f}', 
                  fontsize=9)
        plt.xlabel('')
        plt.ylabel('')
        plt.tick_params(labelsize=7)
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()

# Method 3: Identify and visualize features by type
def categorize_and_plot(df):
    """
    Categorize features and plot them separately
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    # Categorize features
    binary_features = [col for col in numeric_cols if df[col].nunique() <= 2]
    categorical_features = [col for col in numeric_cols 
                           if df[col].nunique() > 2 and df[col].nunique() <= 10]
    continuous_features = [col for col in numeric_cols 
                          if col not in binary_features and col not in categorical_features]
    
    print(f"Binary features ({len(binary_features)}): {binary_features[:5]}...")
    print(f"Categorical features ({len(categorical_features)}): {categorical_features[:5]}...")
    print(f"Continuous features ({len(continuous_features)}): {continuous_features[:5]}...")
    
    # Plot binary features as bar charts
    if binary_features:
        n_binary = len(binary_features)
        fig, axes = plt.subplots(int(np.ceil(n_binary/5)), 5, figsize=(20, int(np.ceil(n_binary/5)*4)))
        axes = axes.flatten() if n_binary > 5 else [axes]
        
        for idx, col in enumerate(binary_features):
            df[col].value_counts().plot(kind='bar', ax=axes[idx], edgecolor='black')
            axes[idx].set_title(f'{col}', fontsize=10)
            axes[idx].tick_params(labelsize=8)
        
        for idx in range(n_binary, len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.savefig('binary_features.png', dpi=150, bbox_inches='tight')
        plt.show()
    
    # Plot categorical features as bar charts
    if categorical_features:
        n_cat = len(categorical_features)
        fig, axes = plt.subplots(int(np.ceil(n_cat/5)), 5, figsize=(20, int(np.ceil(n_cat/5)*4)))
        axes = axes.flatten() if n_cat > 5 else [axes]
        
        for idx, col in enumerate(categorical_features):
            df[col].value_counts().plot(kind='bar', ax=axes[idx], edgecolor='black')
            axes[idx].set_title(f'{col}', fontsize=10)
            axes[idx].tick_params(labelsize=8)
        
        for idx in range(n_cat, len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.savefig('categorical_features.png', dpi=150, bbox_inches='tight')
        plt.show()
    
    # Plot continuous features as histograms
    if continuous_features:
        n_cont = len(continuous_features)
        fig, axes = plt.subplots(int(np.ceil(n_cont/5)), 5, figsize=(20, int(np.ceil(n_cont/5)*4)))
        axes = axes.flatten() if n_cont > 5 else [axes]
        
        for idx, col in enumerate(continuous_features):
            axes[idx].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
            axes[idx].set_title(f'{col}', fontsize=10)
            axes[idx].tick_params(labelsize=8)
        
        for idx in range(n_cont, len(axes)):
            axes[idx].set_visible(False)
        
        plt.tight_layout()
        plt.savefig('continuous_features.png', dpi=150, bbox_inches='tight')
        plt.show()

# Method 4: Create an interactive summary
def create_summary_report(df, output_file='distribution_summary.txt'):
    """
    Generate a text-based summary report
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    with open(output_file, 'w') as f:
        f.write("CREDIT CARD DATASET - DISTRIBUTION SUMMARY\n")
        f.write("=" * 80 + "\n\n")
        
        for col in numeric_cols:
            f.write(f"\nFeature: {col}\n")
            f.write("-" * 40 + "\n")
            f.write(f"Count: {df[col].count()}\n")
            f.write(f"Missing: {df[col].isna().sum()}\n")
            f.write(f"Mean: {df[col].mean():.4f}\n")
            f.write(f"Std: {df[col].std():.4f}\n")
            f.write(f"Min: {df[col].min():.4f}\n")
            f.write(f"25%: {df[col].quantile(0.25):.4f}\n")
            f.write(f"50%: {df[col].quantile(0.50):.4f}\n")
            f.write(f"75%: {df[col].quantile(0.75):.4f}\n")
            f.write(f"Max: {df[col].max():.4f}\n")
            f.write(f"Unique values: {df[col].nunique()}\n")
    
    print(f"Summary report saved to {output_file}")
```

```python
# Method 5: Box plots for all features
def plot_all_boxplots(df, figsize=(20, 60), cols_per_row=5):
    """
    Create box plots for all numeric features to identify outliers
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = len(numeric_cols)
    n_rows = int(np.ceil(n_cols / cols_per_row))
    
    fig, axes = plt.subplots(n_rows, cols_per_row, figsize=figsize)
    axes = axes.flatten()
    
    for idx, col in enumerate(numeric_cols):
        ax = axes[idx]
        df.boxplot(column=col, ax=ax)
        ax.set_title(f'{col}', fontsize=10)
        ax.set_xlabel('')
        ax.tick_params(labelsize=7)
    
    # Hide empty subplots
    for idx in range(n_cols, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('all_boxplots.png', dpi=150, bbox_inches='tight')
    plt.show()

# Method 6: Correlation heatmap
def plot_correlation_heatmap(df, figsize=(20, 18), method='pearson'):
    """
    Create a correlation heatmap for all numeric features
    method: 'pearson', 'spearman', or 'kendall'
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    corr_matrix = df[numeric_cols].corr(method=method)
    
    plt.figure(figsize=figsize)
    sns.heatmap(corr_matrix, 
                cmap='coolwarm', 
                center=0,
                annot=False,  # Set to True if you want to see correlation values
                fmt='.2f',
                square=True,
                linewidths=0.5,
                cbar_kws={"shrink": 0.8})
    plt.title(f'Correlation Heatmap ({method.capitalize()})', fontsize=16, pad=20)
    plt.tight_layout()
    plt.savefig('correlation_heatmap.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    return corr_matrix

# Method 7: Scatter plots for highly correlated features
def plot_high_correlations(df, threshold=0.7, max_plots=20):
    """
    Create scatter plots for pairs of features with high correlation
    threshold: minimum absolute correlation to plot (default 0.7)
    max_plots: maximum number of scatter plots to create
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    corr_matrix = df[numeric_cols].corr()
    
    # Find high correlations (excluding diagonal)
    high_corr_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) >= threshold:
                high_corr_pairs.append({
                    'feature1': corr_matrix.columns[i],
                    'feature2': corr_matrix.columns[j],
                    'correlation': corr_matrix.iloc[i, j]
                })
    
    # Sort by absolute correlation
    high_corr_pairs = sorted(high_corr_pairs, 
                            key=lambda x: abs(x['correlation']), 
                            reverse=True)
    
    print(f"Found {len(high_corr_pairs)} feature pairs with |correlation| >= {threshold}")
    
    if len(high_corr_pairs) == 0:
        print(f"No correlations found above threshold {threshold}")
        return
    
    # Limit number of plots
    pairs_to_plot = high_corr_pairs[:max_plots]
    
    n_plots = len(pairs_to_plot)
    n_cols = 4
    n_rows = int(np.ceil(n_plots / n_cols))
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5*n_rows))
    axes = axes.flatten() if n_plots > 1 else [axes]
    
    for idx, pair in enumerate(pairs_to_plot):
        ax = axes[idx]
        ax.scatter(df[pair['feature1']], df[pair['feature2']], alpha=0.5, s=10)
        ax.set_xlabel(pair['feature1'], fontsize=9)
        ax.set_ylabel(pair['feature2'], fontsize=9)
        ax.set_title(f"Corr: {pair['correlation']:.3f}", fontsize=10)
        ax.tick_params(labelsize=8)
        
        # Add trend line
        z = np.polyfit(df[pair['feature1']].dropna(), 
                      df[pair['feature2']].dropna(), 1)
        p = np.poly1d(z)
        ax.plot(df[pair['feature1']], p(df[pair['feature1']]), 
               "r--", alpha=0.8, linewidth=1)
    
    # Hide empty subplots
    for idx in range(n_plots, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.savefig('high_correlation_scatterplots.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    # Print top correlations
    print("\nTop correlated feature pairs:")
    for pair in pairs_to_plot[:10]:
        print(f"{pair['feature1']} vs {pair['feature2']}: {pair['correlation']:.4f}")

# Method 8: Pairplot for selected features (use carefully with many features)
def plot_pairplot_subset(df, features=None, sample_size=None):
    """
    Create a pairplot for a subset of features
    features: list of feature names to include (if None, uses first 10 numeric features)
    sample_size: number of rows to sample (if None, uses all data)
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    if features is None:
        features = numeric_cols[:10].tolist()
        print(f"No features specified. Using first 10: {features}")
    
    # Sample data if dataset is large
    df_plot = df[features].copy()
    if sample_size and len(df_plot) > sample_size:
        df_plot = df_plot.sample(n=sample_size, random_state=42)
        print(f"Sampled {sample_size} rows for visualization")
    
    print(f"Creating pairplot for {len(features)} features...")
    sns.pairplot(df_plot, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20})
    plt.savefig('pairplot_subset.png', dpi=150, bbox_inches='tight')
    plt.show()

# Method 9: Correlation analysis report
def correlation_report(df, output_file='correlation_report.txt'):
    """
    Generate detailed correlation analysis report
    """
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    corr_matrix = df[numeric_cols].corr()
    
    with open(output_file, 'w') as f:
        f.write("CORRELATION ANALYSIS REPORT\n")
        f.write("=" * 80 + "\n\n")
        
        # High positive correlations
        f.write("HIGH POSITIVE CORRELATIONS (> 0.7):\n")
        f.write("-" * 40 + "\n")
        for i in range(len(corr_matrix.columns)):
            for j in range(i+1, len(corr_matrix.columns)):
                if corr_matrix.iloc[i, j] > 0.7:
                    f.write(f"{corr_matrix.columns[i]} <-> {corr_matrix.columns[j]}: "
                           f"{corr_matrix.iloc[i, j]:.4f}\n")
        
        # High negative correlations
        f.write("\n\nHIGH NEGATIVE CORRELATIONS (< -0.7):\n")
        f.write("-" * 40 + "\n")
        for i in range(len(corr_matrix.columns)):
            for j in range(i+1, len(corr_matrix.columns)):
                if corr_matrix.iloc[i, j] < -0.7:
                    f.write(f"{corr_matrix.columns[i]} <-> {corr_matrix.columns[j]}: "
                           f"{corr_matrix.iloc[i, j]:.4f}\n")
        
        # Features with many high correlations
        f.write("\n\nFEATURES WITH MOST HIGH CORRELATIONS (|r| > 0.7):\n")
        f.write("-" * 40 + "\n")
        high_corr_counts = {}
        for i in range(len(corr_matrix.columns)):
            count = sum(abs(corr_matrix.iloc[i, :]) > 0.7) - 1  # Exclude self
            if count > 0:
                high_corr_counts[corr_matrix.columns[i]] = count
        
        for feature, count in sorted(high_corr_counts.items(), 
                                     key=lambda x: x[1], reverse=True):
            f.write(f"{feature}: {count} high correlations\n")
    
    print(f"Correlation report saved to {output_file}")
```
